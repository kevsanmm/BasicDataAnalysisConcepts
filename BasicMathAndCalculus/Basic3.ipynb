```python
# Charting a linear function in Python using SymPy 
from sympy import * 

x = symbols('x') 
f = 2*x + 1 
plot(f)
```

```python
# Charting an exponential function 
from sympy import * 

x = symbols('x') 
f = x**2 + 1
plot(f)
```

```python
# Declaring a function with two independent variables in Python 
from sympy import * 
from sympy.plotting import plot3d 

x, y = symbols('x y') 
f = 2*x + 3*y 
plot3d(f)
```
